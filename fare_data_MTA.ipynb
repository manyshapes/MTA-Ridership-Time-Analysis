```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
```

```python
# Source: http://web.mta.info/developers/fare.html
def get_data(week_nums):
    url = "http://web.mta.info/developers/data/nyct/fares/fares_{}.csv"
    dfs = []
    for week_num in week_nums:
        file_url = url.format(week_num)
        dfs.append(pd.read_csv(file_url, index_col=0, names=['REMOTE','STATION','FF', 'SEN/DIS', '7-D AFAS UNL','30-D AFAS/RMF UNL','JOINT RR TKT','7-D UNL','30-D UNL','7D-XBUS PASS','TCMC','RF 2 TRIP','RR UNL NO TRADE','TCMC ANNUAL MC','MR EZPAY EXP','MR EZPAY UNL', 'PATH 2-T','AIRTRAIN FF','AIRTRAIN 30-D','AIRTRAIN 10-T','AIRTRAIN MTHLY','STUDENTS','NICE 2-T','CUNY-120','CUNY-60','FF VALUE','FF 7-DAY','FF 30-DAY','R001','WHITEHALL STREET']))
    return pd.concat(dfs)
```

```python
#every saturday week num starts
#year , month, day

#june 2020
week_nums = [200627, 200620, 200613]
june_fares = get_data(week_nums)

#dropping rows
june_fares = june_fares.drop(june_fares.index[[0,2]])

#drop last column
june_df_fares = june_fares.drop('WHITEHALL STREET', axis=1) 
june_df_fares
```

```python
#march 2020
week_nums = [200328, 200321, 200314]
march_fares = get_data(week_nums)

#dropping rows
march_fares = march_fares.drop(march_fares.index[[0,2]])
march_fares

#drop last column
march_df_fares = march_fares.drop('WHITEHALL STREET', axis=1) 
march_df_fares
```
